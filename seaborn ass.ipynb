# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating aesthetically pleasing and informative statistical graphics. Here are five commonly used plots that can be created using Seaborn:

Scatter Plot:

Uses: Scatter plots are used to visualize the relationship between two numerical variables. They help in identifying patterns, correlations, or clusters in the data.

Bar Plot:

Uses: Bar plots are useful for comparing categorical data. They display the distribution or frequency of categories and allow easy comparison between different categories.

Histogram:

Uses: Histograms are employed to visualize the distribution of a single numerical variable. They show how data is distributed across different bins or intervals.

Box Plot:

Uses: Box plots, also known as box-and-whisker plots, are used to display the distribution of numerical data and show the spread, central tendency, and identify outliers.

Heatmap:

Uses: Heatmaps are used to visualize the magnitude of a relationship between two categorical variables. They are particularly useful in showing correlation matrices and highlighting patterns in large datasets.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot(x = "timepoint" , y = "signal", data = fmri)
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot("month" , "year" , "passengers")
```

```python
sns.heatmap(flights_pivot, cmap='viridis' , fmt='d' , annot=True)
```
